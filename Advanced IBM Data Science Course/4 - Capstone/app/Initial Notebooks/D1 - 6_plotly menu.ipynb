### Machine Learning models. 

In this process we will use some machine learning models in order to see how is the performance on each one. 

```python
import plotly.express as px
import pandas as pd

# Datos de ejemplo
data = px.data.iris()

# Crear figura inicial
fig = px.scatter(data, x="sepal_width", y="sepal_length", color="species")

# Obtener nombres de las columnas del dataframe
columnas = list(data.columns)
columnas.remove("species")  # Eliminar "species" de las opciones de eje x e y

# Crear diccionario de botones para el menú desplegable del eje x
buttons_x = []
for columna in columnas:
    button = dict(
        args=[{"x": [data[columna]]}],
        label=columna,
        method="restyle"
    )
    buttons_x.append(button)

# Crear diccionario de botones para el menú desplegable del eje y
buttons_y = []
for columna in columnas:
    button = dict(
        args=[{"y": [data[columna]]}],
        label=columna,
        method="restyle"
    )
    buttons_y.append(button)

# Crear diccionario de botones para el menú desplegable de colores
buttons_color = []
for columna in columnas:
    button = dict(
        args=[{"color": data[columna]}],
        label=columna,
        method="restyle"
    )
    buttons_color.append(button)

# Actualizar la figura con los menús desplegables para el eje x, y y los colores
fig.update_layout(
    updatemenus=[
        dict(
            buttons=buttons_x,
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top",
            active=1
        ),
        dict(
            buttons=buttons_y,
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.3,
            xanchor="left",
            y=1.1,
            yanchor="top",
            active=2
        ),
        dict(
            buttons=buttons_color,
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.5,
            xanchor="left",
            y=1.1,
            yanchor="top",
            active=0
        )
    ]
)

# Mostrar el gráfico
fig.show()
```
